
# 📚 Introdução ao Pandas

O Pandas é uma biblioteca essencial do Python para **análise e manipulação de dados estruturados**, como planilhas e tabelas.

---

## 🧠 O que você pode fazer com Pandas:

- **Carregar dados** de arquivos CSV, Excel, SQL etc.
- **Visualizar rapidamente** os dados com `.head()` e `.info()`
- **Filtrar** linhas e colunas com facilidade
- **Limpar dados** nulos, duplicados e tipos errados
- **Criar e modificar colunas**
- **Agrupar e resumir dados** com `groupby()`
- **Juntar tabelas** (como em banco de dados)
- **Exportar os resultados** para CSV ou Excel
- **Criar dashboards simples** com Streamlit, usando DataFrames

---

## ✅ Principais estruturas do Pandas:

### 1. `Series`  
Uma coluna com rótulos (índice). Exemplo:
```python
import pandas as pd
precos = pd.Series([10, 20, 30], name="Preço")
```

### 2. `DataFrame`  
Uma tabela completa com linhas e colunas. Exemplo:
```python
dados = {
    "produto": ["Camiseta", "Calça", "Boné"],
    "preco": [49.90, 89.90, 29.90]
}
df = pd.DataFrame(dados)
```
